```r
library(dotenv)
# Load environment variables from the .env file
env_file <- normalizePath("../../.env", mustWork = FALSE) # Ensure relative path works
if (!file.exists(env_file)) {
  stop(paste("Environment file not found at:", env_file))
}
load_dot_env(env_file)

# Get the workspace path from the .env file
workspace_path <- Sys.getenv("WORKSPACE_PATH", unset = NA)
if (is.na(workspace_path) || workspace_path == "") {
  stop("WORKSPACE_PATH is not set in the .env file.")
}
```
